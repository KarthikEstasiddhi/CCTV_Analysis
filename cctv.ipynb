```python
import cv2
import torch
import networkx as nx
import matplotlib.pyplot as plt
from ultralytics import YOLO
import time  # Import time module to manage delays

# Load YOLO model (downloaded automatically)
model = YOLO('yolov8n.pt')  # Use the smallest YOLOv8 model for testing

# Initialize Graph
G = nx.Graph()

def update_graph(detections, frame_id):
    """ Update the network graph based on detected persons """
    G.clear()  # Reset the graph for each frame
    
    for i, det in enumerate(detections):
        x1, y1, x2, y2, conf, cls = det
        label = int(cls)
        if label == 0:  # Class 0 = person in COCO dataset
            node_id = f"Person_{i}_{frame_id}"
            G.add_node(node_id, pos=((x1 + x2) // 2, (y1 + y2) // 2))
            
            # Connect to existing nodes if close enough
            for other_node in list(G.nodes):
                if other_node != node_id:
                    ox, oy = G.nodes[other_node]['pos']
                    distance = ((ox - (x1 + x2) // 2) ** 2 + (oy - (y1 + y2) // 2) ** 2) ** 0.5
                    if distance < 100:
                        G.add_edge(node_id, other_node)
                        
def draw_graph():
    """ Draw and update the graph dynamically """
    pos = nx.get_node_attributes(G, 'pos')
    nx.draw(G, pos, with_labels=True, node_size=200, font_size=8, font_color='red', node_color='yellow')

    plt.draw()
    plt.show(block=False)
    plt.pause(1)
    plt.clf()

def show_frame_matplotlib(frame):
    """ Display OpenCV frame using Matplotlib instead of cv2.imshow() """
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for correct colors
    plt.imshow(frame_rgb)
    plt.axis("off")
    plt.show(block=False)
    plt.pause(0.001)
    plt.clf()

def main():
    cap = cv2.VideoCapture('sampleVideo.mp4')  # Load video file

    if not cap.isOpened():
        print("Error: Could not open video file")
        return

    frame_id = 0
    last_frame_time = 0  # Track the last time a frame was displayed
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        
        frame_id += 1
        current_time = cap.get(cv2.CAP_PROP_POS_MSEC) / 1000  # Get the current timestamp in seconds

        # Only process the frame if at least 1 second has passed since the last one
        if current_time - last_frame_time >= 1:
            # Run YOLO detection
            results = model(frame)[0].boxes.data.cpu().numpy()

            # Debugging: Print detection results
            print(f"Frame {frame_id}: Detected {len(results)} objects")
            print(results)  # Print detected objects

            # Update the graph based on detections
            update_graph(results, frame_id)
            
            # Draw detection boxes on the frame
            for x1, y1, x2, y2, conf, cls in results:
                if int(cls) == 0:
                    cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
                    cv2.putText(frame, f'Person {conf:.2f}', (int(x1), int(y1) - 5),
                                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1)

            # Show the frame using Matplotlib instead of OpenCV
            show_frame_matplotlib(frame)

            # Draw graph separately
            draw_graph()

            # Update the time when the frame was shown
            last_frame_time = current_time

    cap.release()
    print("Video processing complete.")

if __name__ == "__main__":
    main()
```
